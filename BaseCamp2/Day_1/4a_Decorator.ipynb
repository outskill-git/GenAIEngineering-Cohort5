**Decorator Function**  
The Function which can act as a wrapper in a generic way  
Can provide add-on features and can be added for any of the functions  

```python
# Define the decorator function as a wrapper function
# the decorator takes a funciton and returns the wrapper function
def deco_fn (func) :

    def wrapper(*args):
        
        print("Additional Functionality - Before ...")
        func(*args)   # call original
        print("Additional Functionality - After ...")
    
    return wrapper
```

```python
# The Additional functionality added by the decorator is added to various functions
# Note the syntax : Simple @
@deco_fn
def Addition (x, y):

    print ("Sum of x & y is : ", str(x+y))

@deco_fn
def Status (st):

    print ("Status : ", st)
```

> Oiriginal function, with its same name is now added with additional features  
> Its callable as itself

```python
# Call to the original functions will work as the original + additional features
Addition (2, 8)

Status ("pass")
```

> Decorator function that will capture time stamp and display lapsed time for a functoinality  
> This can be used like #timeit

```python
import time

def time_lapsed (func):

    # Make the wrapper to capture start and end time
    # print the time lapsed
    def wrapper(*args):
        start_time = time.time()
        result = func(*args) # Call the original function
        end_time = time.time()
        
        print(f"{func.__name__} took {end_time - start_time:.4f} seconds")
        return result
    return wrapper
```

```python
# use it on any function
@time_lapsed
def task (time_sec):

    # Simulate an intense task by delay
    time.sleep (time_sec)

@time_lapsed
def compute (factor) :

    rslt  = 0

    # Long computation
    for idx in range (factor) :

        rslt = rslt + ((factor * 1.2) + (factor + 2.78)) * 1.456
    
    return rslt        
```

```python
# Try with different delay values and check
task (1)

task (4)
```
