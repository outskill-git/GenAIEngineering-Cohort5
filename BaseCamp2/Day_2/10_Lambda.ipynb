### Lambda Function
Short function which consist expression of logic     
No definition, directly used as expresssion and its inputs   
Syntax : 'lambda argument : expression'

```python
import pandas as pd
```

```python
# A list of tuples: (Iterarion, Accuracy)
models = [('I_1', 0.88), ('I_2', 0.92), ('I_3', 0.75), ('I_4', 0.81)]

# Sort by the second element (Accuracy)
# The lambda takes the tuple 'x' and returns 'x[1]' as the sorting key.
# Note that lambda is given as an expression. used inside sort function
models.sort(key=lambda x: x[1], reverse=True)

print(models)
```

```python
text_data = ["  Hello World ", "AI is Cool", "  Python  "]

# Strip whitespace and convert to lower case in one go
clean_data = list(map(lambda s: s.strip().lower(), text_data))

print(clean_data)
```

**Use of lambda in dataframe**  
lambda functions can be quite handy in pandas dataframe, if there are some processing required in the data  
there could be scenario where pandas operations directly could not be possible or be ambiguous  
lambda functions can be used in such scneario

```python
# Data with a field with a data structure as such
# The user action data contains the user events per user
data = {
    'user_id': [101, 102, 103],
    # Column contains a list of dictionaries
    'events': [
        [{'type': 'click', 'time': '10:00'}, {'type': 'view', 'time': '10:05'}],
        [{'type': 'click', 'time': '11:00'}, {'type': 'purchase', 'time': '11:15'}, {'type': 'view', 'time': '11:20'}],
        [{'type': 'scroll', 'time': '12:00'}]
    ]
}
user_data = pd.DataFrame(data)
user_data
```

```python
# Create a new column 'is_purchaser' (Boolean)

# The lambda function checks if ANY item in the list 'x' 
# has the 'type' key equal to 'purchase'.
user_data['is_purchaser'] = user_data['events'].apply(lambda x: any(event['type'] == 'purchase' for event in x))

user_data
```

```python
temp_data = pd.DataFrame({
    'temp': [20, 35, 10, 45],
    'humidity': [50, 90, 20, 30]
})

temp_data
```

```python
# Custom logic for "Threshold Level"
# apply this logic for each row and perform the check.
# Use of expression - similar to comprehensions

temp_data['risk'] = temp_data.apply(
    lambda row: "High" if row['temp'] > 40 or (row['temp'] > 30 and row['humidity'] > 80) else "Low", 
    axis=1
)

temp_data
```

```python
Index_Data = pd.read_csv ('Index.csv', parse_dates=['Date'], date_format="%d-%b-%y")
print (Index_Data.dtypes)
```

```python
Index_Data.sort_values (by='Date', inplace=True)

# Calculate the minimum of the next 5 days' Lows (M_L)
# This is the lowest possible price in the next 5 days.
# Rolling function works as 'Backward looking' - This in combination with shift can be made into 'forward looking'.
Index_Data['next_5_low_min'] = Index_Data['Low'].rolling(window=5).min().shift(-5)

# Calculate the maximum of the next 5 days' Highs (M_H)
# This is the highest possible price in the next 5 days.
Index_Data['next_5_high_max'] = Index_Data['High'].rolling(window=5).max().shift(-5)
# Index_Data.to_csv ('Temp.csv', index=False, float_format='%.2f')
```

```python
Index_Data['trend'] = Index_Data.apply(
    lambda row: (
        'Bullish' if row['next_5_low_min'] > row['Close'] else
        'Bearish' if row['next_5_high_max'] < row['Close'] else
        'Neutral'
    ),
    axis=1 
)

# Index_Data.to_csv ('Temp.csv', index=False, float_format='%.2f')
Index_Data
```
